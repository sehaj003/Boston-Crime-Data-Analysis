```python
correlation_results = {}
for factor in ['UNRATE(%)', 'CONSUMER CONF INDEX', 'PPI-CONST MAT.', 'CPIALLITEMS', 'INFLATION(%)', 'MORTGAGE INT. MONTHLY AVG(%)', 'MED HOUSEHOLD INCOME', 'CORP. BOND YIELD(%)','GDP PER CAPITA' ,'QUARTERLY REAL GDP' ,'QUARTERLY GDP GROWTH RATE (%)']:
    correlation = crime_counts_by_month.corr(merged_df.groupby('MONTH')[factor].sum(), method='pearson')
    correlation_results[factor] = correlation
print("Pearson Correlations:")
for factor, correlation in correlation_results.items():
    print(f"{factor}: {correlation}")
```

The passage of comprehensive legislation in California in 2022 aimed at regulating the ghost gun industry had a significant impact on public safety, particularly in terms of assaults involving deadly weapons. By implementing stricter regulations, the state aimed to curb the availability of unserialized firearms, commonly known as ghost guns, and impose stringent requirements on their sale, transfer, and possession.

Under the new laws:

1. **Strict Regulation of Ghost Guns:** Unfinished frames or receivers, which are essential components for assembling functional firearms, are now subject to the same regulations as fully assembled guns. This means that the sale, transfer, and possession of these unfinished components must be conducted through licensed firearms dealers. Individuals must undergo in-person ID checks, background checks, and waiting periods, similar to those required for fully assembled firearms.

2. **Serial Number Requirement:** Unfinished frames or receivers must have federal serial numbers imprinted by licensed manufacturers. The law prohibits the sale or possession of unserialized firearms, ensuring that all firearms in circulation are traceable and accounted for.

3. **License Requirements:** Anyone manufacturing more than three firearms in a calendar year or using 3D printers or CNC milling machines to produce firearms must obtain a firearm manufacturer's license. This requirement ensures that individuals engaged in firearm production are regulated, undergo background checks, and maintain manufacturing records.

4. **Personal Use Limitations:** Individuals without a firearm manufacturer's license are limited to producing up to three firearms per year for personal use only. These firearms must comply with specific regulations, including engraving or affixing unique serial numbers and passing background checks.

5. **Prohibition of Unserialized Firearms:** Possession, sale, or transfer of unserialized firearms, including both completed and unfinished frames or receivers, is generally unlawful. This restriction ensures that unserialized firearms are not available to individuals who could potentially use them for criminal purposes.

By enforcing these laws, California has significantly reduced the availability of unserialized firearms, making it harder for individuals to obtain deadly weapons without proper oversight. This, in turn, contributes to a decrease in assaults involving untraceable and unregulated firearms, enhancing public safety and ensuring law enforcement can effectively track and investigate firearm-related incidents.

```python
import pandas as pd
from scipy import stats

def find_outliers(dataframe, z_score_threshold=2):
    # Create an empty DataFrame to store the outliers
    outliers_df = pd.DataFrame()

    # Iterate over each column in the DataFrame
    for column in dataframe.columns:
        column_data = dataframe[column]
        z_scores = stats.zscore(column_data)
        outliers = dataframe[abs(z_scores) > z_score_threshold]
        outliers_df = pd.concat([outliers_df, outliers])

    return outliers_df

# Load your data into a DataFrame (replace 'your_data.csv' with your data file)

# Set a Z-Score threshold (adjust as needed)
z_score_threshold = 2

# Call the function to find outliers in every column
outliers_result = find_outliers(clean_crimeDataFrame, z_score_threshold)

# Display the rows containing outliers for each column
print(outliers_result)
```
